```python
def count_residues(pdb_file_path):
    residues = set()

    with open(pdb_file_path, 'r') as file:
        for line in file:
            if line.startswith("ATOM") or line.startswith("HETATM"):
                chain_id = line[21]
                res_seq = line[22:26].strip()
                insertion_code = line[26].strip()
                residue_id = (chain_id, res_seq, insertion_code)
                residues.add(residue_id)

    return len(residues)

# Example usage
if __name__ == "__main__":
    pdb_path = "8xiz__B1_T2B4K5-L.pdb"  # Replace with your file path
    num_residues = count_residues(pdb_path)
    print(f"Number of unique residues: {num_residues}")
```
