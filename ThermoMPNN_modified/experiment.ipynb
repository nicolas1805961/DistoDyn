```python
import pandas as pd

df = pd.read_csv("data_all/testing/mega_test.csv")

# extract prefix before first dot (e.g. 3DKM from 3DKM.pdb_wte)
df["name_prefix"] = df["name"].str.split(".", n=1).str[0]

# assign x per prefix
df["x"] = pd.factorize(df["name_prefix"])[0]

# assign y per row within prefix
df["y"] = df.groupby("name_prefix").cumcount()

# create new_name
df["new_name"] = "protein_" + df["x"].astype(str) + "_" + df["y"].astype(str)

# cleanup
df.drop(columns=["name_prefix", "x", "y"], inplace=True)

# save updated dataframe
df.to_csv("pandora_with_new_name.csv", index=False)

print("Saved pandora_with_new_name.csv")
```

```python
from glob import glob
import os
import torch
import pickle
import pandas as pd
import math

def wt_sort_key(f):
    basename = os.path.basename(f)
    wt_name = basename.split("_model")[0].replace("distogram_", "")
    # assume wt_name is like 'protein_32'
    num = int(wt_name.split("_")[-1])
    return num

def normalize_distogram(disto_logits):
    # convert to probabilities
    disto = torch.softmax(disto_logits, dim=-1)

    # Option 1: simple centering
    disto = disto - disto.mean(dim=-1, keepdim=True)

    # Option 2: scaling to unit variance per sample
    disto = disto / (disto.std(dim=-1, keepdim=True) + 1e-6)

    return disto

for split in ['train', 'validation', 'test']:
    # only load rows needed to save memory
    #df = pd.read_csv(fname, usecols=["dG_ML", "mut_type", "WT_name", "aa_seq", "wt_seq"])
    df = pd.read_csv('pandora.csv', usecols=["ddG_ML", "mut_type", "new_name", "aa_seq", "wt_seq", "split"])
    df = df[df["split"] == split].copy()

    # remove unreliable data and more complicated mutations
    df = df.loc[df.ddG_ML != '-', :].reset_index(drop=True)
    df = df.loc[~df.mut_type.str.contains("ins") & ~df.mut_type.str.contains("del") & ~df.mut_type.str.contains(":"), :].reset_index(drop=True)

    # load splits produced by mmseqs clustering
    #with open(self.cfg.data_loc.megascale_splits, 'rb') as f:
    #    splits = pickle.load(f)  # this is a dict with keys train/val/test and items holding FULL PDB names for a given split

    wt_names = sorted({name.rsplit("_", 1)[0] for name in df["new_name"].values},key=lambda x: int(x.split("_")[1]))

    print(len(wt_names))

    #self.distogram_files = glob(os.path.join('/pasteur/appa/scratch/nportal/boltz/stability_prediction/results_wt', '**', '**', "distogram*.pkl"), recursive=True)
    # Get all distogram files (unique)
    all_files = list(set(glob(os.path.join('new_wt_pt', '**', '**', "distogram*.pkl"), recursive=True)))

    # Keep only files whose name matches a WT in this split
    # Keep only files whose basename matches exactly a WT in this split
    distogram_files = [
        f for f in all_files
        if os.path.basename(f).startswith("distogram_") and
        os.path.basename(f).split("_model")[0].replace("distogram_", "") in wt_names
    ]

    distogram_files = sorted(distogram_files, key=wt_sort_key)

    print(f"Number of distogram files for split '{split}':", len(distogram_files))

    for fpath in distogram_files:
        bin_edge_path = fpath.replace("distogram", "bin_edge")
        print(bin_edge_path)
        with open(fpath, "rb") as f:
            data = pickle.load(f)
        disto = torch.from_numpy(data['distogram']['logits']).float().unsqueeze(0)  # add batch dim
        bin_edges = torch.from_numpy(data['distogram']['bin_edges']).float().unsqueeze(0)  # add batch dim
        disto = normalize_distogram(disto)
        torch.save(disto, fpath.replace(".pkl", ".pt"))
        torch.save(bin_edges, bin_edge_path.replace(".pkl", ".pt"))
```
