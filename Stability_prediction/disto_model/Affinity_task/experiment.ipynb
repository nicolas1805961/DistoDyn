```python
import h5py

def print_affinities(filepath):
    """Iterate over keys in an HDF5 file and print the 'affinity' value if it exists."""
    with h5py.File(filepath, 'r') as f:
        for key in f.keys():
            group = f[key]
            print(group.keys())
            print(f"\n[{key}]")
            if "affinity" in group:
                affinity = group["affinity"][()]
                affinityType = group["affinityType"][()]
                print(f"  affinity: {affinity}")
                print(f"  affinityType: {affinityType}")
            else:
                print("  'affinity' key not found")

if __name__ == "__main__":
    filepath = "affinity_data.h5"  # replace with your file path
    print_affinities(filepath)
```
